```python
# Carga de tabla histórica generada
import pandas as pd

df = pd.read_parquet("../data/historico_us.parquet")
df.head()
```

```python
#Revisión de calidad de los datos
import pandas as pd

# Inicializar lista de chequeos
verificacion = []

# 1. Duplicados
duplicados = df.duplicated(subset=["state", "report_date"]).sum()
verificacion.append({
    "Chequeo": "Duplicados por estado y fecha",
    "Resultado": duplicados,
    "Interpretación": "✅ Sin duplicados" if duplicados == 0 else "⚠️ Revisar registros duplicados"
})

# 2. Nulos por columnas (solo si > 0)
nulos = df.isna().sum()
for col, val in nulos[nulos > 0].items():
    verificacion.append({
        "Chequeo": f"Valores nulos en '{col}'",
        "Resultado": val,
        "Interpretación": "⚠️ Considera imputar o revisar"
    })

if nulos.sum() == 0:
    verificacion.append({
        "Chequeo": "Valores nulos en todas las columnas",
        "Resultado": 0,
        "Interpretación": "✅ No hay valores nulos"
    })

# 3. Negativos en columnas numéricas clave
negativos = df[["confirmed", "deaths", "recovered", "active"]].lt(0).sum()
for col, val in negativos.items():
    verificacion.append({
        "Chequeo": f"Valores negativos en '{col}'",
        "Resultado": val,
        "Interpretación": "✅ Sin negativos" if val == 0 else "⚠️ Revisar inconsistencias"
    })

# Mostrar tabla final
pd.DataFrame(verificacion)
```
